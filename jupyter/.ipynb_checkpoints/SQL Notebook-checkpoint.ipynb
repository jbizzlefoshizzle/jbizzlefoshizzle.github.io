```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine, func
from pandas.plotting import table
```

```python
rds_connection_string = "postgres:jbm713177@localhost:5432/beer_db" #establish connection to postgres database
engine = create_engine(f'postgresql://{rds_connection_string}') #use sqlalchemy
engine.table_names() #check to see if connection works
```

```python
coordinates = pd.read_sql_query('SELECT * FROM brewery_coordinates', con=engine)
coordinates.head()
```

```python
medals = pd.read_sql_query('SELECT id, brewery, COUNT(medal) as medal_count \
    FROM beer_winners b \
    GROUP BY b.id', con=engine)
medals.head()
```

```python
medals_and_coordinates = pd.read_sql_query('\
    SELECT b.id, b.name, b.rating, b.review_count, b.lat, b.long, \
    w.medal \
    FROM brewery_coordinates b \
    LEFT JOIN beer_winners w \
    ON b.name = w.brewery', 
                                           con=engine)
medals_and_coordinates
```
