```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.greatamericanbeerfestival.com/the-competition/winners/'
browser.visit(url)
```

```python
table_list = []
years_list = ['2017', '2016', '2015', '2014', '2013', 
              '2012', '2011', '2010', '2009', '2008', 
              '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000', '1999']
```

```python
#start loop
for year in years_list:
    
    #Find dropdown based on current year and click
    dropdown = browser.find_by_id('year')
    dropdown.click()

    #Click on link
    links_found = browser.find_link_by_partial_text(year)
    links_found.click()
    
    #select html using splinter
    html = browser.html
    #capture entire html using BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    current_table = soup.find("table", id="winners_table")
    #append current_table to table_list
    table_list.append(current_table)
```

```python
# print one of the Beautiful Soup Table Objects
print(current_table.prettify())
```

```python
#create empty list to store table data
l = []

for table in table_list:
    
    table_rows = table.find_all('tr')
    for tr in table_rows[1:]:
        td = tr.find_all('td')
        row = [tr.text for tr in td]
        l.append(row)
    
#convert list 'l' to a DataFrame
complete_df = pd.DataFrame(l, columns=['Medal', 'Beer Name', 'Brewery', 'City', 'State', 'Category', 'Year'])
complete_df.head(15)
```
